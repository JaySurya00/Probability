On the game show Let's Make a Deal, hosted by Monty Hall, a contestant choose one of the three closed doors, two of which have a goat behind them and one which has a car. Monty knows where the car is, then opens one of the two remaining doors. The door he opens always has a goat behind them (he never reveals the car). If he has a choice, then he picks a door at random with equal probabilities. Monty then offers the contestant the option of switching to the other unopened door. If the contestant's goal is to win the car, should he/she switch door?

```python
import random

def monty_hall(n=3, m=1, trials=1):
    stay_wins=0
    switch_wins=0

    for _ in range(trials):
        doors= list(range(n))
        car= random.choice(doors)
        choice= random.choice(doors)

        #Monty open m goat door

        possible_opens= [d for d in doors if d!=choice and d!=car]
        opened= set(random.sample(possible_opens, m))

        unopened_doors= [d for d in doors if d not in opened and d!=choice ]

        # Stays
        if choice==car:
            stay_wins+=1
        
        # Switch
        if random.choice(unopened_doors)==car:
            switch_wins+=1

    print(f"Stay wins= {stay_wins/trials}")
    print(f"Switch wins= {switch_wins/trials}")

monty_hall(3, 1, 100000)


```
