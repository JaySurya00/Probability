```python
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
```

Suppose you perform an experiment of tossing a coin n times. The probability of getting a head is p. You have to calculate the probability of success(getting a head) X times in the experiment.

Simulation using numpy

```python
n= 10
p=0.5
size= 10000
samples= np.random.binomial(n, p, size)
print(samples)
```

```python
plt.hist(samples, bins=range(n+2), density=True, edgecolor='black')
plt.xlabel("Number of success")
plt.ylabel("Probability")
plt.title("Simulation of Binomial Distribution")
plt.show()
```

```python
x= np.arange(0, n+1)
pmf= binom.pmf(x, n, p)
plt.hist(samples, bins=range(n+2), density=True, edgecolor='black', label= 'Simulation')
plt.plot(x, pmf, 'o-', label= 'Theoretical PMF')
plt.xlabel("Number of success")
plt.ylabel("PMF")
plt.title("Theoretical PMF")
plt.show()
```
